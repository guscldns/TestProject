### 5) 무한루프는 continue, break 같이 쓴다.
- continue, break : 둘다 중단!
 - continue : 지금 하는 루프는 중단하고 다음 걸로 넘어가겠다

        print(1)
        print(2)
        continue
        print(3)
        >> 3 실행하지 않고 다시 1,2로 돌아가게 됨
        >> 1,2 1,2 1,2 1,2, ......

 - break : 중단하고 끝낸다.(종료)

        print(1)
        print(2)
        break
        print(3)
        >> 3 실행안되고 1,2 끝
