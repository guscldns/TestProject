```python
import cv2
import os
import pandas as pd
import numpy as np

# 주어진 데이터프레임에서 bbox 좌표와 파일명을 가져옵니다.
bbox_df = label_total[['bbox', 'file_name']]

# 빈 리스트를 생성하여 모든 디스크립터를 저장할 준비를 합니다.
all_descriptors = []

# 이미지 폴더의 이미지를 하나씩 처리합니다.
for index, row in bbox_df.iterrows():
    # 이미지 파일을 읽어옵니다.
    image_path = f"C:/Users/guscldns/Downloads/dataset/labeled_images/{row['file_name']}"
    image = cv2.imread(image_path)

    # bbox 좌표를 가져옵니다.
    bbox = row['bbox']
    x, y, width, height = map(int, bbox)

    # bbox 영역에서 물고기를 추출합니다.
    fish_roi = image[y:y+height, x:x+width]

    # SIFT 객체 생성
    sift = cv2.SIFT_create()

    # 키 포인트와 디스크립터 추출
    keypoints, descriptors = sift.detectAndCompute(fish_roi, None)

    # 현재 이미지의 디스크립터를 모든 디스크립터 리스트에 추가
    all_descriptors.append(descriptors)

# 모든 디스크립터를 하나의 Numpy 배열로 병합
all_descriptors_array = np.vstack(all_descriptors)

# Numpy 배열을 데이터프레임으로 변환
descriptor_df = pd.DataFrame(all_descriptors_array)

# 결과 데이터프레임 출력
print(descriptor_df)
```

```python
import cv2
import numpy as np
import pandas as pd

def extract_features(image):
    # 이미지를 그레이스케일로 변환합니다.
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 색상 특성 추출
    mean_color = np.mean(image, axis=(0, 1)) # RGB 색상 평균 계산
    std_color = np.std(image, axis=(0, 1))  # RGB 색상 표준 편차 계산
    max_color = np.max(image, axis=(0, 1))  # RGB 색상 최댓값 계산
    mode_color = np.argmax(np.bincount(image.reshape(-1, 3).astype(int).flatten()), axis=0)  # RGB 색상 최빈값 계산

    # 밝기 특성 추출
    mean_brightness = np.mean(gray)
    std_brightness = np.std(gray)

    # 명암 특성 추출
    laplacian = cv2.Laplacian(gray, cv2.CV_64F).var()
    hist = cv2.calcHist([gray], [0], None, [256], [0, 256]) # 명암 히스토그램

    # 크기 특성 추출
    size = image.shape[0] * image.shape[1]

    # 추출한 특성을 딕셔너리로 저장하여 반환합니다.
    features = {'mean_color': mean_color, 'std_color': std_color, 'max_color': max_color, 'mode_color': mode_color,
                'mean_brightness': mean_brightness, 'std_brightness': std_brightness, 'laplacian': laplacian,
                'histogram': hist, 'size': size}
    return features


# 주어진 데이터프레임에서 bbox 좌표와 파일명을 가져옵니다.
bbox_df = label_total[['bbox', 'file_name']]

# 새로운 데이터프레임을 저장할 리스트를 생성합니다.
new_df = []

# 각 이미지에 대해 물고기의 특성을 추출합니다.
for index, row in bbox_df.iterrows():
    # 이미지 파일을 읽어옵니다.
    image_path = f"C:/Users/guscldns/Downloads/dataset/labeled_images/{row['file_name']}"
    image = cv2.imread(image_path)

    # bbox 좌표를 가져옵니다.
    bbox = row['bbox']
    x, y, width, height = map(int, bbox)

    # bbox 영역에서 물고기를 추출합니다.
    fish_roi = image[y:y+height, x:x+width]

    # 물고기의 특성을 추출하여 수치화합니다.
    feature = extract_features(fish_roi)

    # 새로운 데이터프레임에 특성과 파일명을 추가합니다.
    new_row = {'file_name': row['file_name']}
    new_row.update(feature)
    new_df.append(new_row)

# 새로운 데이터프레임을 생성합니다.
new_df = pd.DataFrame(new_df)
```

```python
train['image_id']= train['image_id'].fillna('0').astype(int)
train['category_id']= train['category_id'].fillna('-1').astype(int)
train['area']= train['area'].fillna('0.0')
train['bbox']= train['bbox'].fillna('0000')
train['bbox_x'] = train['bbox'].apply(lambda x: x[0])
train['bbox_y'] = train['bbox'].apply(lambda x: x[1])
train['bbox_w'] = train['bbox'].apply(lambda x: x[2])
train['bbox_h'] = train['bbox'].apply(lambda x: x[3])
train['iscrowd']= train['iscrowd'].fillna('-1').astype(int)
train['id_y']= train['id_y'].fillna('0').astype(int)
train['name']= train['name'].fillna('없음')

train = train.drop(columns='bbox', axis=1)
train
```

```python
# 데이터 프레임 내의 이미지 파일 경로 수정
train['image_path'] = 'C:/Users/guscldns/Downloads/dataset/train/' + train['file_name']

# 데이터를 train 및 test로 분할
train_data, test_data = train_test_split(train, test_size=0.2, random_state=42)
```

```python
# 모델 저장 경로 정의
model_path = f"./tmp/{uuid.uuid4().hex}-automm_custom"

# 모델 학습
predictor = MultiModalPredictor(label='name', path=model_path, eval_metric='f1_macro').fit(train_data=train_data, time_limit=30)
```
