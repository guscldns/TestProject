```python
ebola = pd.read_csv("country_timeseries.csv")
ebola2 = pd.melt(ebola, id_vars = ['Date', 'Day'])
ebola2

# var_name, value_name 안써주면 자동으로 var_name은 variable, value_name은 value로 바꿔줌
```

```python
# variable을 생존 기준으로 나누고 싶음

ebola_split = ebola2['variable'].str.split('_')
# .str : 문자형으로 변환
# .split('_') : 문자형데이터를 _ 기준으로 잘라라

ebola_split
```

```python
state = ebola_split.str[0] # state에 ebola_split를 문자화한 0번쨰 값 가져옴
country = ebola_split.str[1] # country에 ebola_split를 문자화한 1번쨰 값 가져옴

ebola2['State'] = state # State 열 만들고 그 안에 state 데이터 채움
ebola2['Country'] = country # Country 열 만들고 그 안에 country 데이터 채움

del ebola2['variable'] # 필요없어진 variable 값 삭제

ebola2
```
