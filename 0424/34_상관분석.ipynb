```python
from sklearn.datasets import load_diabetes # 당뇨병 데이터
import seaborn as sns # 데이터 분석 연습용
import pandas as pd # 데이터 분석
import scipy.stats as stats # 상관계수를 분석하기 위해서
```

```python
diabetes = load_diabetes() # dic
```

```python
diabetes.keys()
```

```python
data, target = diabetes['data'], diabetes['target']
```

```python
data
```

```python
target # 당뇨병 진행도
```

```python
pd.DataFrame(data)
```

```python
df = pd.DataFrame(data, columns = diabetes['feature_names'])
df['target'] = target
df
```

```python
X = df['age']
Y = df['target']

stats.pearsonr(X, Y) # 상관계수, p-value
# (statistic=0.18788875071891983, pvalue=7.055686149116415e-05)
# 0.18788875071891983 : 상관계수
# pvalue=7.055686149116415e-05 : 7의 -5승 = 거의 0에 가깝다(상관관계가 없을 확율이 0에 가깝다)
```

```python
X = df['sex']
Y = df['target']

stats.pearsonr(X, Y) # 상관계수 , p-value 
# 0.04306199845160542 : 상관계수가 0.04 = 별로 상관없다 >> 0.05보다 작으면 데이터는 맞으나 믿을 수는 없다
# pvalue=0.36642929465198143 (상관관계가 없을 확률 = 0.36%)
```

```python
X = df['bmi']
Y = df['target']

stats.pearsonr(X, Y) # 상관계수 , p-value 
# 0.5864501344746887 =  0.5이상 : 강한 상관관계
# pvalue=3.466006445167151e-42 = -42승 : 거의 0
```

```python
df.corr() # 한번에 상관계수를 구할 수 있음
# 맨 아래것만 보면 됨
```

```python
sns.heatmap(df.corr()) # sns.heatmap : 상관계수 그래프
# 색이 밝을 수록 상관관계가 높다는 것을 알 수 있음
```
