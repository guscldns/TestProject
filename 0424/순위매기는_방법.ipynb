```python
df = pd.read_excel("성적 처리.xlsx", sheet_name = 'Sheet1')
df
```

```python
df.groupby('반').mean()
# groupby : 묶는 것
# mean : 평균값
```

```python
df['총합'] = df['국어']+ df['영어']+df['수학']+df['사회']+df['과학']
df
```

```python
df.iloc[:,2:]
# iloc[:, 2:] > 앞이 행, 뒤에가 열
# [:, ] > 열 처음부터 끝까지 (슬라이싱)
# [, 2:] > 2행 부터 끝까지
```

```python
df.iloc[:,2:].sum(axis = 0)
# sum(axis = 0) : axis = 0 (행방향)으로 합쳐라
```

```python
df.iloc[:,2:].sum(axis = 1)
# sum(axis = 1) : axis = 1 (열방향)으로 합쳐라
```

```python
df['총합']=df.iloc[:,2:].sum(axis = 1)
df # 총합 열 만들어서 데이터 넣기
```

```python
df['평균']= df['총합']/5
df
```

```python
df['순위'] = df['평균'].rank(method='average')
df 
 # df['평균'].rank(method='average') 
 # df['평균'] : 기준값
 # method='average' : 동점처리 = 평균(소수점 랭킹 생김)
 # 점수가 크면 순위도 높아처 정렬다시 필요
```

```python
df['순위'] = df['평균'].rank(method='average', ascending=False)
df 
# ascending=False: 내림차순
```

```python
df['순위'] = df['평균'].rank(method='min', ascending=False)
df
# method='min' : 동점 = 그룹 내 최소 순위 부여(아래 예시에서는 3등이 2명)
```

```python
df['순위'] = df['평균'].rank(method='max', ascending=False)
df
# method='max' : 동점 = 그룹 내 최대 순위 
# (아래 예시에서는 동점자 둘다 4등, 3등 없음)
```

```python
df['순위'] = df['평균'].rank(method='first', ascending=False)
df
# method='first' : 데이터 상에 제일 먼저 나온 점수에 순위부여
# 예시에서는 홍길동이 먼저 나와서 3위 다니엘이 4위
```

```python
df['순위'] = df['평균'].rank(method='dense', ascending=False)
df
# method='dense': 'min'과 같음, 동점 = 최소 순위
```
