```python
import pandas as pd

fish = pd.read_csv("fish.csv")
fish # length , weight : 문제지, class : 정답지
```

```python
data = fish[['length','weight']].to_numpy()
data
```

```python
target = fish['class'].to_numpy()
target
```

```python
from sklearn.model_selection import train_test_split #훈련, 시험 데이터 분할 위해 필요
from sklearn.neighbors import KNeighborsClassifier # 이웃 가져옴, 모델, 학습위함

# 훈련문제집, 시험문제집, 훈련정답지, 시험정답지 = train_test_split(문제집, 정답지, test = 비율)
# 위의 순서대로 적어줘야 함, 순서가 중요. 이름은 크게 중요하지 않음
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size = 0.25)
#  test_size = 0.25 는 75 대 25비율인데 이 비율은 기본 값이라 이 비율로 하려면 따로 적지 않아도 됨

knn = KNeighborsClassifier() # 모델 생성
knn.fit(train_input, train_target) # 학습 시 훈련데이터 문제집, 훈련데이터 문제집으로 학습
```

```python
knn.score(test_input, test_target)
# 시험 문제, 시험 답지로 평가했는데도 1,0 (100점)
```
