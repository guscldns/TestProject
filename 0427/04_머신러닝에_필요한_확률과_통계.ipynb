```python
import pandas as pd
fish =pd.read_csv("fish.csv")
fish # 1 : 도미/ 0: 빙어
```

```python
bream = fish[fish['class']==1]
bream
```

```python
smelt = fish[fish['class']==0]
smelt
```

```python
bream_length = list(bream['length'])
bream_weight = list(bream['weight'])
smelt_length = list(smelt['length'])
smelt_weight = list(smelt['weight'])
```

```python
bream_length # 시리즈로 가져옴 >> 리스트 처리해주면 좋음 
# 현재 데이터에서는 병렬 계산 방지용으로 리스트화 필요
```

```python
print(bream_length)
print(bream_weight)
```

```python
print(smelt_length)
print(smelt_weight)
```

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
```

```python
import matplotlib.pyplot as plt

plt.scatter(smelt_length, smelt_weight)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.legend(['bream','smelt'])
```

```python
# 문제지(2차원 행렬) : 길이, 무게
# 정답지(1차원 행렬) : 도미, 빙어 (생선종류)
# 도미 35마리, 빙어 14마리

length = bream_length + smelt_length
print(length) 
```

```python
weight = bream_weight + smelt_weight
print(weight)
```

```python
# 문제지(현재 리스트, 행렬로 바꿔줘야 함)
data = []

for i in range(49):
    data.append([length[i],weight[i]])

data
```

```python
# 정답지 (현재 리스트, 행렬로 바꿔줘야 함)
target = [1] * 35 + [0] * 14
print(target)
```

```python
import numpy as np

data = np.array(data) # array : 행렬로 변환
data
```
