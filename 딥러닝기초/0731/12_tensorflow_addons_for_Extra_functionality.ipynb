# tensorflow addons API for Extra functionality

TensorFlow SIG Addons은 core TensorFlow에서 사용할 수 없는 새로운 기능을 지원한다.

TensorFlow는 기본적으로 다수의 연산자, 레이어, 측정항목, 손실, 옵티마이저 등을 지원하지만, ML과 같이 빠르게 변화하는 분야에는 core TensorFlow에 통합할 수 없는 흥미롭고 새로운 개발 요소가 많다.  
출처 : https://www.tensorflow.org/addons/api_docs/python/tfa

tensorflow addons에서 지원하는 기능은 다양하지만, 그 중 주요 항목은 아래와 같다.
1. tfa.activation *
2. tfa.callbacks
3. tfa.layers (-> normalizations)
4. tfa.losses
5. metrics
6. optimizers  *

그 중에서도 주로 사용하는 activation, optimizer만 다뤄보도록 하자.

## 2) optimizer

### AdamW
Adam의 유일한 단점이 일반화에 약하다는 것인데,
decoupled weight decay를 추가함으로써 일반화 성능을 더 높인 optimizer

```python
tfa.optimizers.AdamW(
    weight_decay: Union[FloatTensorLike, Callable],
    learning_rate: Union[FloatTensorLike, Callable] = 0.001,
    beta_1: Union[FloatTensorLike, Callable] = 0.9,
    beta_2: Union[FloatTensorLike, Callable] = 0.999,
    epsilon: tfa.types.FloatTensorLike = 1e-07,
    amsgrad: bool = False,
    name: str = 'AdamW',
    **kwargs
)
```

### SGDW
decoupled weight decay를 추가함으로써 일반화 성능을 더 높인 optimizer

```python
tfa.optimizers.SGDW(
    weight_decay: Union[FloatTensorLike, Callable],
    learning_rate: Union[FloatTensorLike, Callable] = 0.001,
    momentum: Union[FloatTensorLike, Callable] = 0.0,
    nesterov: bool = False,
    name: str = 'SGDW',
    **kwargs
)
```

### LazyAdam
LazyAdam은 희소 업데이트를 보다 효율적으로 처리하는 Adam 옵티마이저의 변형  
(원래 Adam 알고리즘은 각 훈련 가능한 변수에 대해 두 개의 moving-average accumulator를 유지하는데, 현재 배치에 나타나는 희소 변수 인덱스에 대한 이동 평균 누산기만 업데이트)

```python
tfa.optimizers.LazyAdam(
    learning_rate: Union[FloatTensorLike, Callable] = 0.001,
    beta_1: tfa.types.FloatTensorLike = 0.9,
    beta_2: tfa.types.FloatTensorLike = 0.999,
    epsilon: tfa.types.FloatTensorLike = 1e-07,
    amsgrad: bool = False,
    name: str = 'LazyAdam',
    **kwargs
)
```
