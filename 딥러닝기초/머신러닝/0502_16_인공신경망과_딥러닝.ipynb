```python
from sklearn.datasets import load_iris # 연습데이터
from sklearn.model_selection import train_test_split # 시험데이터
from tensorflow import keras # 인공신경망
from sklearn.preprocessing import StandardScaler # 표준화

iris = load_iris() # 변수에 데이터 넣어줌
data, target = iris['data'], iris['target'] # data, target에 각각 데이터 넣어줌

train_input, test_input, train_target, test_target = train_test_split(data, target)

# 인공신경망이든 딥러닝이든 결국 선형데이터 > 표준화 필요

ss = StandardScaler() # 데이터 표준화
train_scaled = ss.fit_transform(train_input) # 훈련데이터 표준화
test_scaled = ss.transform(test_input) # 테스트 데이터 표준화

model = keras.Sequential() # 인공신경망을 만들기 위한 도화지
model.add(keras.layers.Dense(3, activation='softmax', input_shape=(4,)))
# input_shape 는 튜플로 만들어 줘야함 

model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# 어떻게 학습할지 정해줌

# 앞에 변수 넣어주면 모든 학습을 기록해줌 ↓
hist = model.fit(train_scaled, train_target, epochs = 500)
```

```python
import matplotlib.pyplot as plt

plt.plot(hist.history['loss']) # 오차에 대해 어떻게 학습했는지 보여줌
```

```python
model.evaluate(test_scaled, test_target) # keras 평가하려면 evaluate 써야 함
```

```python
from sklearn.datasets import load_iris # 연습데이터
from sklearn.model_selection import train_test_split # 시험데이터
from tensorflow import keras # 인공신경망
from sklearn.preprocessing import StandardScaler # 표준화

iris = load_iris() 
data, target = iris['data'], iris['target'] 
train_input, test_input, train_target, test_target = train_test_split(data, target)

ss = StandardScaler() 
train_scaled = ss.fit_transform(train_input) 
test_scaled = ss.transform(test_input) 

model = keras.Sequential()
model.add(keras.layers.Dense(8, activation='relu', input_shape=(4,))) 
# 앞에 Dense = 8 : 은닉층 / activation >> relu 로 바꿔줌
# 딥러닝 ↓
model.add(keras.layers.Dense(3,activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])

hist = model.fit(train_scaled, train_target, epochs = 500)
```

```python
import matplotlib.pyplot as plt

plt.plot(hist.history['loss'])
```

```python
model.evaluate(test_scaled, test_target) # 평가
```
