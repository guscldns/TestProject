```python
import pandas as pd
from bs4 import BeautifulSoup
import time
import requests

total = []

url = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EA%B1%B4%EA%B0%95&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=46&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=1")
html = BeautifulSoup(url.text)

news = html.find('ul', class_='list_news').find_all('li',class_='bx')

for i in news:
    title = i.find('a', class_='news_tit')
    contents= i.find('a', class_='api_txt_lines dsc_txt_wrap')

    total.append([title, contents])

pd.DataFrame(total, columns=["제목","요약"])
```
