```python
import json

url = requests.get("https://project-team.upbit.com/api/v1/disclosure?region=kr&per_page=20")
url.text
# html = BeautifulSoup(url.text) 안함 
# 해당 페이지의 데이터 보면 html아님 >> text로 그냥 가져옴
# {}가 보임 = 딕셔너리
data = json.loads(url.text)
data
# json.load(url.text): 제이슨 처럼 보이지만 실제 제이슨은 아님 >> 문자열을 실제 제이슨으로 변환시켜주는 함수
# data 변수에 json.loads(url.text)로 url.text 있으니 4번째 줄 지워도 됨
# 딕셔너리 안에 {딕셔너리{딕셔너리[리스트]}} 형태임
#{data : {'more': True,'offset': 574,'posts: [{...'text': 제목',...}]} 형태
# 먼저 맨 앞의 딕셔너리, 그 다음 딕셔너리 가져와야 함
```

```python
import json

url = requests.get("https://project-team.upbit.com/api/v1/disclosure?region=kr&per_page=20")
data = json.loads(url.text)
data['data']['posts']
# ['data'] :첫번째 딕셔너리
# ['posts] : 첫번째 딕셔너리 안의 두번쨰 딕셔너리
# 풀었더니 리스트 나옴 [{id:...}]
# 리스트 해제 필요 >> for 문 사용
```

```python
import json

url = requests.get("https://project-team.upbit.com/api/v1/disclosure?region=kr&per_page=20")
data = json.loads(url.text)

for i in data['data']['posts']:
    print(i) # >> 리스트를 풀어서 안의 딕셔너리 가져옴
    # 제목은 text 키 값에 있음
```

```python
import json

url = requests.get("https://project-team.upbit.com/api/v1/disclosure?region=kr&per_page=20")
data = json.loads(url.text)

for i in data['data']['posts']:
    print(i['text'])
    # i['text'] : i의 text 키 값에 접근하겠다
```
