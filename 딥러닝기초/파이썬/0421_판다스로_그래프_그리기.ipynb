```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset("tips") # tips 데이터 가져오기(연습용)
tips
```

```python
tips['total_bill'].plot.hist() 
# plot : 그리라는 것
# hist : 히스토그램 작성
```

```python
tips[['total_bill','tip']].plot.hist() 
# [['total_bill','tip']] : 여러개에 접근할 때는 항상 하나로 묶어 줘야 함
```

```python
tips[['total_bill','tip','size']].plot.hist(alpha = 0.5) 
# alpha : 투명도 / 0.5 = 50%
```

```python
tips[['total_bill','tip','size']].plot.hist(alpha = 0.5, bins = 30) 
# bins : 막대기 개수
# 색 진하기에 따라 겹친 정도 알 수 있음
```

```python
tips['tip'].plot.kde()
# kde: 밀도 함수, 그래프 안쪽을 다 색칠하면 1 (밀도가 다 합쳐지면 1이라서)
```

```python
tips.plot.scatter(x='total_bill', y='tip', c ='size')
# scatter : 산점도
```

```python
tips.plot.scatter(x='total_bill', y='tip', c ='sex', s=tips['size']*10)
# s=tips['size'] : 점 크기, * 해서 점 크기 키울 수 있음
# 점 크기가 크고 색이 진할 수록 많은 데이터
```

```python
tips.plot.hexbin(x='total_bill', y='tip')
# hexbin : 육각형모양 산점도
```

```python
tips.plot.hexbin(x='total_bill', y='tip', gridsize=10)
# gridsize : 육각형 크기 (숫자가 작을 수록 크기가 커짐)
```

```python
tips.plot.box()
# box : 박스 그래프 그릴 수 있음
# 단, 숫자로된 데이터만 가능
```
