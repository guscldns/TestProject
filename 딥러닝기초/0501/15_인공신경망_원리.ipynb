```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([2,4,6,8]) # 행렬로 해줘야 함
y = np.array([30,75,55,90])

a1 = 0
b = 0 
# 수가 너무 커서 올라가니 작은 값 곱해줘야 함 
lr = 0.005  # 곱해줄 작은 값
# 경사하강법 ↓
for i in range(1, 3001):
    y_hat = a1*x1+b
    # y_hat : 예측선
    error = y_hat - y
    # error : 오차
    
    # 미분값(2 * error * x1) 
    a1_diff = sum(2 * error * x1)
    b_diff = sum(2 * error *1)
    # 너무 오래걸리니 합쳐줌

    a1 = a1 - a1_diff * lr # 작은 값 곱해줌
    b = b - b_diff * lr

    print("{}회 학습/ 기울기 {} / 절편 {}".format(i, a1, b))
# 수가 점점 작아지면서 일정 수가 반복되는 것을 보임 / 기울기 8.0... 절편 22.49...
```

```python
plt.scatter(x1,y)
plt.plot(x1, a1 * x1+b, 'red')
# 그래프 그려보면 가장 좋은 직선 그려짐을 알 수 있음
# 가장 좋은 직선 : 가장 오차가 적은 직선
# 해당 직선을 통해 예측 가능
```
