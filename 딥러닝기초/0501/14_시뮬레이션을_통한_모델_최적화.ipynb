```python
from sklearn.datasets import load_wine # 예제 와인데이터 가져옴
import pandas as pd # 어떻게 나왔는지 보기 위해 판다스 가져옴

wine = load_wine()
```

```python
df = pd.DataFrame(wine['data']) # 와인데이터 확인
df
```

```python
df = pd.DataFrame(wine['data'], columns = wine['feature_names']) # 열 이름 가져옴
df
```

```python
df = pd.DataFrame(wine['data'], columns = wine['feature_names'])
df['target'] = wine['target'] # target 열 추가, 데이터는 와인의 target
df
```

```python
wine['target_names']
# class인 것으로 봐서 종류인 것으로 추측
# 그림은 특성이 많아서 무리
# 데이터 표준화는 랜덤포레스로 
```

```python
from sklearn.ensemble import RandomForestClassifier # 램덤포레스트 가져옴
from sklearn.model_selection import GridSearchCV # GridSerch 가져옴

data, target = wine['data'], wine['target']
```

```python
rf = RandomForestClassifier() 
# 랜덤포레스트 모델 , ()안에 매개변수 넣을 수 있음
# () 안에  n_estimators=100이 기본값 : 결정트리를 100개 만들겠다 > 수정하고 싶으면 변경 가능

dic = {'n_estimators' : range(100, 201, 10), "max_depth" : range(1,11)}
# n_estimators : 결정트리 , range(100, 201, 10) : 100개부터 200개 까지 10개 씩
# max_depth : 내려갈 수 있는 최대 층, range(1,11): 1층부터 10층까지만 가겠다
gs = GridSearchCV(estimator= rf, param_grid= dic,cv=3)
# GridSearchCV(estimator= 모델, param_grid= dic, cv=n)
# estimator= : 매개변수
# dic = 딕셔너리 형태의 옵션값, 
# dic 만들 때 딕셔너리 내 원하는만큼 옵션 추가할 수 있음
# cv=n : 교차검증

gs.fit(data, target) # 학습
```

```python
# 100, max(1~10)
# 110, max(1~10) >> 이런 식으로 학습한 것

gs.best_params_ # 최적값
```

```python
result = gs.cv_results_  # 결과 정리
df = pd.DataFrame(result) # 데이터프레임화
df.to_excel("result.xlsx") # 엑셀 파일로 다운로드
# 엑셀에서 1등이 제일 좋은 성적을 낸 모델
```

```python
df
```

```python
best = gs.best_estimator_

# best_estimator_ :1등인 모델을 변수에 넣을 수 있음
# 여기서는 best 라는 변수에 넣음
# best 변수의 모델로 훈련, 평가 할 수 있음
```

```python
best.score(data, target) # best모델로 평가
# 정확도 1.0(100%) 나옴
```
