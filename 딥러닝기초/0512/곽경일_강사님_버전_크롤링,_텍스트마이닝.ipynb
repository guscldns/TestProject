```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import time
import pandas as pd
```

```python
total =[]
for n in tqdm(range(1,3992,10)):
    url = requests.get(f'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%95%8C%ED%8C%8C%EC%BD%94&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=23&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={n}')
    html = BeautifulSoup(url.text, 'lxml')

    try:
        contents = html.find("ul", class_ = 'list_news').find_all('li', class_ ='bx')
        len(contents)
    except:
        continue

    for i in contents:
        title = i.find("a", class_='news_tit')['title']
        surmary = i.find('div',class_='dsc_wrap').text
        total.append([title, surmary])
    time.sleep(0.6)
```

```python
len(total)
```

```python
df = pd.DataFrame(total, columns =['제목','요약'])
df
```

```python
t = ''
    # t에 제목 공백 요약 공백인 한 문장을 만들겠다 ↓
for i, j in zip(df['제목'], df['요약']):
    t += i
    t += " "
    t += j
    t += " "


len(t)
```

```python
import re # 큰 문장에서 내가 원하는 패턴만 가져오는 것

#re.findall("[a]", t)  # t에서 a 라는 패턴을 찾겠다
#re.findall("[abc]", t) # t에서 a나 b나 c 인 패턴을 찾겠다
#re.findall("[a-z]", t)  # t에서 영어소문자인 패턴을 찾겠다 # a-z : 모든 영어 소문자
#re.findall("[a-z]+", t)  # t에서 영어소문자가 하나라도 들어간 단어 패턴을 찾겠다 # [a-z] + : 모든 영어 소문자가 들어간 것들 찾기
#re.findall("[a-zA-Z]+", t) # A-Z : 모든 대문자
#re.findall("[a-zA-Z0-9]+", t)# 0-9 : 모든 숫자
nouns = re.findall("[a-zA-Z0-9가-힣]+", t) # 가-힣 : 모든 한글
len(nouns)
```

```python
 # konlpy : 한국 자연어 처리 // re 하면 조사를 자동으로 없애주고 명사만 가져와 줌

nouns # 단어를 포함한 것을 다 가져오는데 조사도 포함한 것도 가져와짐 (알파코'는' 등)
```

```python
# 빈도 분석 (nouns에서 교육이라는 단어가 몇 번이나 나올까?)
# 빈 딕셔너리 생성
dic = {}
for i in nouns:
    if i not in dic:
        dic[i] = 1 # 없는 단어가 발견될 때 마다 빈 딕셔너리에 넣겠다
    else:
        dic[i] += 1 # 딕셔너리 안에 단어가 있으면 1 추가 >> 빈도 분석
dic
```
