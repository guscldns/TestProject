```python
def conv_block(x, fillters, kenrnel_size):
    x = tf.keras.layers.Conv2D(fillters, kenrnel_size, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    return x

def darknet19(input_shape = (224, 224, 3), num_classes = 1000):
    inputs = tf.keras.Input(shape = input_shape)

    x = conv_block(inputs, 32, 3)
    x = tf.keras.layers.MaxPooling2D(pool_size = 2, strides = 2)(x)

    x = conv_block(x, 64, 3)
    x = tf.keras.layers.MaxPooling2D(pool_size = 2, strides = 2)(x)

    x = conv_block(x, 128, 3)
    x = conv_block(x, 64, 1)
    x = conv_block(x, 128, 3)
    x = tf.keras.layers.MaxPooling2D(pool_size = 2, strides = 2)(x)

    x = conv_block(x, 256, 3)
    x = conv_block(x, 128, 1)
    x = conv_block(x, 256, 3)
    x = tf.keras.layers.MaxPooling2D(pool_size = 2, strides = 2)(x)

    x = conv_block(x, 512, 3)
    x = conv_block(x, 256, 1)
    x = conv_block(x, 512, 3)
    x = conv_block(x, 256, 1)
    x = conv_block(x, 512, 3)
    x = tf.keras.layers.MaxPooling2D(pool_size = 2, strides = 2)(x)

    x = conv_block(x, 1024, 3)
    x = conv_block(x, 512, 1)
    x = conv_block(x, 1024, 3)
    x = conv_block(x, 512, 1)
    x = conv_block(x, 1024, 3)

    x = conv_block(x, 1000, 1)
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = tf.keras.layers.Dense(1000, activation = 'softmax')(x)

    model = tf.keras.Model(inputs = inputs , outputs = x)
    return model
```

```python
def yolo_v2(input_shape):
    inputs = tf.keras.layers.Input(shape = input_shape)
    x = tf.keras.layers.Conv2D(filters = 32, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling2D(pool_size=2, strides = 2)(x)
    x = tf.keras.layers.Conv2D(filters = 32, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling2D(pool_size=2, strides = 2)(x)
    x = tf.keras.layers.Conv2D(filters = 128, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 64, kernel_size= 1, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 128, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling2D(pool_size=2, strides = 2)(x)
    x = tf.keras.layers.Conv2D(filters = 256, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 128, kernel_size= 1, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 256, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling2D(pool_size=2, strides = 2)(x)
    x = tf.keras.layers.Conv2D(filters = 512, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 256, kernel_size= 1, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 512, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 256, kernel_size= 1, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    reorg = tf.keras.layers.Conv2D(filters = 512, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(reorg)
    x = tf.keras.layers.MaxPooling2D(pool_size=2, strides = 2)(x)
    x = tf.keras.layers.Conv2D(filters = 1024, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 512, kernel_size= 1, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 1024, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 512, kernel_size= 1, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 1024, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 1024, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Conv2D(filters = 1024, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    reorg_1 = reorg[:, ::2, ::2, :]
    reorg_2 = reorg[:, ::2, 1::2, :]
    reorg_3 = reorg[:, 1::2, ::2, :]
    reorg_4 = reorg[:, 1::2, 1::2, :]
    concat = tf.concat([reorg_1, reorg_2, reorg_3, reorg_4], axis=-1, name='concat')
    concatted = tf.concat([concat, x], axis = -1)
    x =  tf.keras.layers.Conv2D(filters = 1024, kernel_size= 3, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(concatted)
    x =  tf.keras.layers.Conv2D(filters = 1024, kernel_size= 1, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    model = tf.keras.Model(inputs = inputs, outputs = x)
    return model
```

```python
def conv_b(x, fillters, kenrnel_size):
    x = tf.keras.layers.Conv2D(fillters, kenrnel_size, strides = 1, padding = 'same', activation= tf.keras.layers.LeakyReLU(alpha=0.1))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    return x


def Yolo_v2(input_shape = (416, 416, 3)):
    inputs = tf.keras.Input(shape = input_shape)

    x = conv_b(inputs, 32, 3)

    x = tf.keras.layers.MaxPooling2D(pool_size = 2, strides = 2)(x)
    x = conv_b(x, 64, 3)

    x = tf.keras.layers.MaxPooling2D(pool_size=2, strides = 2)(x)
    x = conv_b(x, 128, 3)
    x = conv_b(x, 64, 1)
    x = conv_b(x, 128, 3)

    x = tf.keras.layers.MaxPooling2D(pool_size=2, strides = 2)(x)
    x = conv_b(x, 256, 3)
    x = conv_b(x, 128, 1)
    x = conv_b(x, 256, 3)

    x = tf.keras.layers.MaxPooling2D(pool_size=2, strides = 2)(x)
    x = conv_b(x, 512, 3)
    x = conv_b(x, 256, 1)
    x = conv_b(x, 512, 3)
    x = conv_b(x, 256, 1)
    x = conv_b(x, 512, 3)

    x_1 = x[:, :13, :13, :]
    x_2 = x[:, :13, 13:, :]
    x_3 = x[:, 13:, :13, :]
    x_4 = x[:, 13:, 13:, :]

    reorg = tf.concat([x_1, x_1, x_1, x_1], axis = -1, name = 'reorg')

    x = tf.keras.layers.MaxPooling2D(pool_size=2, strides = 2)(x)
    x = conv_b(x, 1024, 3)
    x = conv_b(x, 512, 1)
    x = conv_b(x, 1024, 3)
    x = conv_b(x, 512, 1)
    x = conv_b(x, 1024, 3)
    x = conv_b(x, 1024, 3)
    x = conv_b(x, 1024, 3)

    x = tf.concat([reorg, x], axis = -1, name = 'route')

    x = conv_b(x, 1024, 3)
    x = conv_b(x, 125, 1)

    model = tf.keras.Model(inputs = inputs , outputs = x)
    return model
```
