```python
import re
```

```python
f = open("코로나.txt", 'rt')
covid = f.read()
f.close

print(covid)
```

```python
re.findall("[가-힣]", covid)
# re.findall(패턴, 문자열) : 문자열에서 패턴을 찾는다
# [가-힣] : 모든 한글
```

```python
re.findall("[가-힣]+", covid)
# [가-힣]+ : 1글자 이상
# + = 1개 이상
```

```python
re.findall("[가-힣]{2}", covid)
# [가-힣]{2} : {2} 2 글자
# {n} : n번 반복
# {m,n} : m부터 n까지 반복
```

```python
re.findall("[가-힣]{2,}", covid)
# {2,} : 2개 이상
```

```python
re.findall("[a-zA-Z]+", covid)
# [a-zA-Z] : 모든 영어
```

```python
re.findall("[0-9]+", covid)
# [0-9] : 모든 숫자
```

```python
re.findall("[\d]+", covid)
# \d : 숫자
```

```python
re.findall("[\D]+", covid)
# \D : 숫자가 아닌것
```

```python
re.findall("[\w+19]", covid)
# \w+19 : 어떤 문자가 1개만 있고 끝이 19로 끝난다 (\w + 19)
# [\w+19] : \w 문자열이 1개이상 있고 1이랑 9가 있다 (\w+  19)
```

```python
re.findall("\w+19", covid)
```

```python
# 코로나19만 찾고 싶다

re.findall("코\w+19", covid)
# 코\w+19 : 코로시작하는 문자이며 끝에 19로 끝난다
```
