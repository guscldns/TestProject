```python
import pandas as pd

ebola=pd.read_csv("country_timeseries.csv")
ebola
```

```python
122-ebola.count()#
```

```python
ebola.fillna(5) # fillna (): 결측치를 채운다
```

```python
ebola.mean()
```

```python
ebola.fillna(ebola.mean()) 
# 시계열 데이터에서는 평균 값으로 하면 데이터가 이상해짐
```

```python
ebola.fillna(method ='ffill')
# fillna(method ='ffill') : 앞에 있는 값으로 채운다 
# Cases_Liberia는 앞에 있는 값이 없어서 NaN 유지됨
```

```python
ebola.fillna(method ='bfill')
# fillna(method ='bfill') 뒤에 있는 값으로 채운다
# Cases_Liberia 밑에 있는 데이터는 뒤에 값이 없어서 못채움
```

```python
ebola.interpolate()
# interpolate() 흐름에 따라 누락값을 채우는 기능
# 흐름을 모르면 결측치가 채워지지 않음
```

```python
# 실무에서 쓰는 방법
ebola2 = ebola.interpolate() # 우선 흐름에 따라 채움
ebola2 = ebola2.fillna(method='ffill') # 앞의 데이터로 채움
ebola2 = ebola2.fillna(method='bfill') # 뒤의 데이터로 채움
ebola2
```

```python
from numpy import nan

dates = ['12/1/2016', '12/3/2016', '12/4/2016', '12/10/2016']
pd.to_datetime(dates) # to_datetime() : '/ '써도 컴퓨터가 날짜로 인식

s = pd.Series([1, nan, nan, 10], index = dates)
s
```

```python
s.interpolate() 
# 흐름에 따라 채웠더니 4. 7 채움, 날짜에 따라 넣는것이 더 효과적
```

```python
from numpy import nan

dates = ['12/1/2016', '12/3/2016', '12/4/2016', '12/10/2016']
pd.to_datetime(dates)
dates = pd.to_datetime(dates)

s=pd.Series([1, nan, nan, 10], index = dates)


s.interpolate(method = 'time')
# method='time': 시간으로 값을 채워라
```
