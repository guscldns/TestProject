```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = requests.get("https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=%EB%A1%9C%EB%98%90")
# URL 페이지 가져올때는 최대한 많이 눌러본다 >> 속성값이 추가되는 경우 많음
# 
url
# 200번대 : 문제없음 / 400번대 : 존재하지 않는 페이지 / 500 번대 : 로그인처리, 잠금
```

```python
url.text
```

```python
html = BeautifulSoup(url.text)
html
```

```python
html.find('span', class_='f_red')
# html에서 span 태그의 f_red라는 클래스 속성값을 가져와라
# 1064회 : text
```

```python
html.find('span', class_='f_red').text
```

```python
html.find('span', class_='f_red').text.replace("회","")
# text.replace("회","") : text에서 회 >> 빈칸으로 바꿔라
```

```python
current = int(html.find('span', class_='f_red').text.replace("회",""))
current
```

```python
html.find("div", class_='lottonum')
# div 태그의 lottonum 속성값을 찾음
```

```python
html.find("div", class_='lottonum').find_all('span')
# 그 안에서 find_all('span) : span이라는 값 모두 찾아라
```

```python
lotto = html.find("div", class_='lottonum').find_all('span')

del lotto[-2] # 밑에서 2번째 지워라
del lotto[-2] # 지운 것에서 밑에서 2번째 지워라

lotto # 번호부분만 남음

# find_all() : 리스트로 가져옴
```

```python
lotto = html.find("div", class_='lottonum').find_all('span')

del lotto[-2] # 밑에서 2번째 지워라
del lotto[-2]

box = [] # 박스 = 리스트
for i in lotto:
    box.append(int(i.text)) # 박스에 text 부분만 정수로 바꿔서 가져와라
    # <span class="ball bg_ball1">3</span> 에서 text : 3 (<>와 <> 사이의 값)

box
```
