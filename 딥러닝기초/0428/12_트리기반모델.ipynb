```python
from sklearn.datasets import load_iris # 사이킷런 : 연습용 데이터 제공

iris = load_iris()
data, target = iris['data'], iris['target']
```

```python
# 연산이 필요없는 모델이라서 시험데이터만 생성

from sklearn.model_selection import train_test_split

train_input, test_input, train_traget, test_target = train_test_split(data, target)
```

```python
from sklearn.tree import DecisionTreeClassifier
# sklearn.tree : 트리모델
# DecisionTreeClassifier : 결정트리 분류

dt = DecisionTreeClassifier() # 결정트리 모델
dt.fit(train_input, train_traget) # 학습
dt.score(test_input, test_target) # 평가
# 끝남 연산이 없어서 코드가 간단하다
```

```python
# 어떻게 학습했나 자세히 보자
from sklearn.tree import plot_tree # plot_tree 가져옴
```

```python
plot_tree(dt)
```

```python
# 사이즈 키우면
import matplotlib.pyplot as plt

plt.figure(figsize = (10,10)) # figsize = (10,10) :가로 10, 세로 10으로 보겠다

plot_tree(dt)
# x[2] >= 2.45 : x의 두번째 열이 2.45보다 크거나 같은가.. 등으로 해석 
```

```python
# 직관적으로 보자
import matplotlib.pyplot as plt

plt.figure(figsize = (12,12)) 
plot_tree(dt, feature_names=iris['feature_names'], filled= True)
```

```python
# 과대적합을 막으려면
from sklearn.model_selection import train_test_split
train_input, test_input, train_traget, test_target = train_test_split(data, target)

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=3) 
# max_depth 의 값을 줘서 그 값 이상으로 진행하지 못하게 막는다
# 여기서는 깊이 3까지 가라 

dt.fit(train_input, train_traget) 
dt.score(test_input, test_target)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,10)) 
plot_tree(dt, feature_names=iris['feature_names'], filled= True)
# 첫번째 질문 제외 3층까지 간 것을 알 수 있음, 너무 층수를 낮게 하면 학습을 다 하기전에 끝날 수도 있음
```

```python
# 랜덤포레스트 
from sklearn.datasets import load_iris # 사이킷런 : 연습용 데이터 제공

iris = load_iris()
data, target = iris['data'], iris['target']

from sklearn.model_selection import train_test_split

train_input, test_input, train_traget, test_target = train_test_split(data, target)

from sklearn.ensemble import RandomForestClassifier # 램덤포레스트
# RandomForestClassifier : 랜덤포레스트 분류

rf = RandomForestClassifier()
rf.fit(train_input, train_traget)
rf.score(test_input, test_target)
# 10개의 결정트리를 가져오는 거라 어떻게 했는지 자세히 보기는 어려움
# 정확도 : 1.0(100%)
```
