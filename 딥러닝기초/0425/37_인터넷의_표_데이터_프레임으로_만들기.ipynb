```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# 페이지 숫자에서 맨뒤를 검사하면 
#<td class="pgRR">
#				<a href="/sise/sise_market_sum.naver?&amp;page=41">맨뒤
#				<img src="https://ssl.pstatic.net/static/n/cmn/bu_pgarRR.gif" width="8" height="5" alt="" border="0">
#				</a>
#				</td>

#'<a href="/sise/sise_market_sum.naver?&amp;page=41">맨뒤'에서 page=41,41페이지까지 존재함을 알 수 있음

html.find('td', class_= 'pgRR').find('a')
```

```python
# 41이라는 데이터만 가져오려면 어디 들어있는지 알아야 함 
# href = << href 안에 있음 알 수 있음

html.find('td', class_= 'pgRR').find('a')['href']
```

```python
# 41은 = 이후에 있음을 알 수 있음
html.find('td', class_= 'pgRR').find('a')['href'].split('=')
# split('=') : = 을 기준으로 잘라줘라
```

```python
html.find('td', class_= 'pgRR').find('a')['href'].split('=')[-1]
# = 을 기준으로 나눠진 리스트의 뒤에서 1번재 값을 가져와라
```

```python
kospi_page = int(html.find('td', class_= 'pgRR').find('a')['href'].split('=')[-1])
```

```python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?&page=1")
html = BeautifulSoup(url.text)
kospi_page = int(html.find('td', class_= 'pgRR').find('a')['href'].split('=')[-1]) # 삽입
# 컴퓨터가 41페이지까지 있음을 알게 됨

# 페이지 넘길 수 있도록 for 문 사용
total = [] 

for n in range(1, kospi_page +1): # 1부터  kospi_page +1 까지 
    url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?&page={}".format(n))
    # page에 {} 하고 format함수에 n 넣어줌
    html = BeautifulSoup(url.text)

    table = html.find('table', class_='type_2')
    table = pd.read_html(str(table))[0]

    table = table[table['종목명'].notnull()]

    del table['N']
    del table['토론실']
    
    total.append(table) # total에 table 값 넣어줘라
    time.sleep(1) # 1초 정도 쉬어라
    
    print("{}페이지 저장 완료...".format(len(total))) 
# 41페이지까지 저장함(대기 필요)
```

```python
len(total) # total의 데이터 길이 
```

```python
total[0] # >> 1번째 페이지 값/ total[1] : 2번째 페이지 갑
```

```python
kospi = pd.concat(total, ignore_index= True)
# pd.concat() 여러개의 데이터프레임을 하나의 데이터 프레임으로 변경 
#pd.concat(total)의 데이터프레임을 하나의 데이터 프레임으로 변경
# ignore_index : 행번호 다시 맞춤
kospi
```

```python
kospi = pd.concat(total, ignore_index= True)
kospi.to_excel("kospi.xlsx") # 엑셀로 저장
kospi
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page=1")
html = BeautifulSoup(url.text)

kosdaq_page = int(html.find('td',class_='pgRR').find('a')['href'].split('=')[-1])

total = []

for i in range(1, kosdaq_page+1):
    url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={}".format(i))
    html = BeautifulSoup(url.text)

    table = html.find('table', class_='type_2')
    table = pd.read_html(str(table))[0]
    table = table[table['종목명'].notnull()]

    del table['N']
    del table['토론실']

    total.append(table)
    time.sleep(1)
    print("{} 페이지 추가중 ...".format(len(total)))
```
