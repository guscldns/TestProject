```python
import pandas as pd
```

```python
# 날씨 정보
# survey(예제)의 데이터 불러오기

person = pd.read_csv("survey_person.csv")
site = pd.read_csv("survey_site.csv")
survey = pd.read_csv("survey_survey.csv")
visited = pd.read_csv("survey_visited.csv")
```

```python
# 날씨를 측정한 사람의 정보
person
```

```python
# 측정한 위치의 정보
site
```

```python
# 날씨 정보
survey
```

```python
# 언제 방문해서 측정했는지 정보
visited
```

```python
# merge로 합치기

merge1 = visited.merge(site, left_on = "site", right_on = "name")

merge1

# site, name의 값이 같아서 지워도 됨
```

```python
del merge1 ['name']
merge1
```

```python
merge2 = survey.merge(person, left_on = 'person', right_on = 'ident')
# merge2가 없어도 합친 데이터 확인은 가능
# merge2 라는 변수가 있어야 다른 데이터 작업이 쉬움

merge2
```
