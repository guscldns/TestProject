문제 17) 팩토리얼 값을 출력하는 코드를 작성해보세요.
* n의 팩토리얼은 n!이라고 표시합니다. 이는 1부터 n까지의 정수를 곱하는 단순한 연산입니다. 예를 들어 5!는 5X4X3X2X1 즉, 120입니다.

n = int(input("n : "))


```python
n = int(input("n! : "))

def fac(n):
    if n==1:
        return 1
    else:
        return n*fac(n-1)
print(fac(n))
```
