```python
import pandas as pd
# 연습용 데이터 파일 업로드
```

```python
df1 = pd.read_csv("concat_1.csv")
df2 = pd.read_csv("concat_2.csv")
df3 = pd.read_csv("concat_3.csv")
# 연습용 데이터 불러오기
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# df1, df2, df3 : ABCD 열이름이 똑같음

pd.concat([df1, df2, df3])

# concat : 여러개의 데이터 프레임을 하나의 데이터 프레임으로 합치는 함수
# 기본 열이름 기준으로 합침
```

```python
concat1 = pd.concat([df1, df2, df3])
# concat1 이라는 변수에 데이터 넣어놓음
# concat1 = pd.concat([df1, df2, df3], axis = 0) 이라고 되어 있는 것
# axis= 0 는 숨겨져 있음
concat1
```

```python
concat1.loc[0]
# loc : [] 안에 해당하는 행을 가져와라
```

```python
concat1 = pd.concat([df1, df2, df3],ignore_index=True)
# ignore_index : 오름차순 정렬 / 기준이 열이름이라 행번호가 오름차순 정렬
concat1
```

```python
concat1 = pd.concat([df1, df2, df3], axis = 1)
# concat : 데이터 합쳐라
# axis = 1 : 행 이름 기준으로 합치기 ↔ axis = 0 : 열 이름 기준
concat1
```

```python
concat2 = pd.concat([df1, df2, df3], axis = 1)
concat2
# concat2에 행이름 기준으로 정렬한 데이터 넣음
```

```python
concat2['A']
# A라고 되어 있는 열의 데이터 확인
```

```python
concat2 = pd.concat([df1, df2, df3], axis = 1, ignore_index=True)
concat2
# ignore_index : 오름차순 정렬/ 행번호가 기준이라 열이름이 오름차순 정렬
# 정렬하면 0부터 숫자로 정렬
```

```python
df2.index 
# 행 이름 확인
```
