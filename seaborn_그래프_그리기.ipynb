```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(Data='tips', x='total bill', y='tips')
```

```python
sns.kdeplot(tips['total_bill'], tips['tip'], shade=True)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
# lmplot : 산점도 그래프
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
# jointplot : 1줄로 히스토그램과 산점도를 한번에 보여줌
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
# hex : 육각형으로 보여줌. 색이 진할수록 많은 값
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
# boxplot : 상자 그래프
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
# 색상변경 가능 >> 단변형 > 다변형으로 그래프 그릴 수 있음
```

```python
sns.violinplot(x='sex', y='total_bill', data=tips)
```

```python
sns.violinplot(x='time', y='total_bill', data=tips, hue='sex')
# 색 변경 가능 > 다변화
```

```python
sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
# 반반 나눠서 볼 수도 있음
```

```python
sns.pairplot(tips)
# 비교할 수 있는 것들은 모두 비교해줌
```

```python
sns.pairplot(tips, hue='sex')
# 색상변경 가능
```

```python
anscombe = sns.load_dataset("anscombe")
sns.lmplot(x='x', y='y', data=anscombe)
```

```python
sns.lmplot(x='x', y='y', data=anscombe, col = 'dataset')
# 집단별로 데이터그려줌
```

```python
sns.lmplot(x='x', y='y', data=anscombe, col = 'dataset', col_wrap = 2)
# col_wrap = 2: 최대 2열까지 표현해라
```

```python
import matplotlib.pyplot as plt

facet = sns.FacetGrid(tips, col="time") # col="time" : 시간을 기준으로 나눠달라
facet.map(plt.scatter, "total_bill", "tip") # x축 : total_bill / y축 : tip
```

```python
import matplotlib.pyplot as plt

facet = sns.FacetGrid(tips, col="time", hue='sex') # 성별로 색변경
facet.map(plt.scatter, "total_bill", "tip")
```

```python
import matplotlib.pyplot as plt

sns.set_style("whitegrid") # 그래프 배경 whitegrid으로 바꿔라

facet = sns.FacetGrid(tips, col="time", hue='sex') # 성별로 색변경
facet.map(plt.scatter, "total_bill", "tip")
```

```python
import matplotlib.pyplot as plt

sns.set_style("whitegrid") 

facet = sns.FacetGrid(tips, col="time", hue='sex', row='smoker') 
# col : 행/ row : 열
# row='smoker' : 열은 smoker 기준으로 작성하라 
facet.map(plt.scatter, "total_bill", "tip")
```
