```python
import matplotlib.pyplot as plt # 가시화 대표적 라이브러리
import seaborn as sns # 데이터분석연습용(연습용 데이터, 가시화)
import pandas as pd
```

```python
# 앤스컴 : 평균과 표준편차, 상관계수, 회귀계수
anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
anscombe[anscombe['dataset'] == 'I'] 
# anscombe에 접근하는데 anscombe['dataset']에 'I'(로마숫자 1)이 있는 데이터에 접근하고 싶다
```

```python
# data에 해당 값 나눠서 넣어주기
data1 = anscombe[anscombe['dataset'] == 'I'] 
data2 = anscombe[anscombe['dataset'] == 'II'] 
data3 = anscombe[anscombe['dataset'] == 'III'] 
data4 = anscombe[anscombe['dataset'] == 'IV'] 
```

```python
print(data1.mean()) # mean : 평균값
print(data2.mean())
print(data3.mean())
print(data4.mean())
```

```python
print(data1.std()) # std : 표준편차
print(data2.std())
print(data3.std())
print(data4.std())
```

```python
fig = plt.figure() # fugure : 그림을 그리기 위한 메인 도화지

ax1 = fig.add_subplot(2,2,1) # 도화지에 자리 할당 (서브도화지)
ax2 = fig.add_subplot(2,2,2) # (2,2,1) : (행, 열, 번째자리) >> 2행 2열 1번째 자리
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
# 산점도 그리기
ax1.scatter(data1['x'], data1['y'])
ax2.scatter(data2['x'], data2['y'])
ax3.scatter(data3['x'], data3['y'])
ax4.scatter(data4['x'], data4['y'])

fig # 데이터 보여주세요
```

```python
# 선그래프 그리기
ax1.plot(data1['x'], data1['y'])
ax2.plot(data2['x'], data2['y'])
ax3.plot(data3['x'], data3['y'])
ax4.plot(data4['x'], data4['y'])

fig # 데이터 보여주세요
```

```python
# 선그래프 그리기 + 색변경
ax1.plot(data1['x'], data1['y'], 'red')
ax2.plot(data2['x'], data2['y'], 'red')
ax3.plot(data3['x'], data3['y'], 'red')
ax4.plot(data4['x'], data4['y'], 'red')

fig # 데이터 보여주세요
```

```python
# 그래프에 제목설정
ax1.set_title("Data1")
ax2.set_title("Data2")
ax3.set_title("Data3")
ax4.set_title("Data4")

fig
```

```python
# 메인 타이틀 설정
fig.suptitle("Main Data")
fig
```

```python
# 제목 붙어있는 것 떨어뜨림
fig.tight_layout()
fig
```
