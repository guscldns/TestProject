```
doc = data['article_text'].iloc[1]
doc
```

```
# 1. 문장으로 쪼개기
from nltk.tokenize import sent_tokenize

sent_list = sent_tokenize(doc)
sent_list
```

```
# 2. 데이터 전처리 
toeknized_sents = []

for sentence in sent_list:
    # 토큰화
    tokens = word_tokenize(sentence)

    sentence = []
    for word in tokens:
        # 영어를 제외한 숫자, 특수 문자 등은 전부 제거. 모든 알파벳은 소문자화
        token = re.sub(r'[^a-zA-z\s]', '', word).lower()
        # r(raw string): 이스케이프 코드 여러개 쓰기 불편하거나 문자열이 깨지거나 할 때 그 문자열을 고정해주는 방법 
        if token not in stop_words and token:# and token : 토큰이 있어야 돌아간다
            sentence.append(token)
    toeknized_sents.append(sentence)
toeknized_sents
```

```
toeknized_sents
```

```
# 3. 문장 벡터
embedding_dim = 100
zero_vector = np.zeros(embedding_dim)

# 단어 벡터의 평균으로부터 문장 벡터를 얻는다.
vect_list = []
for sentence in toeknized_sents:
    if len(sentence) != 0:
        sent_vect = sum([glove_dict.get(word, zero_vector) for word in sentence])/len(sentence)
        # glove_dict: glove에서 가져옴 / word2vec이면 word2vec 쓰면 됨
    else:
        sent_vect = zero_vector
    vect_list.append(sent_vect)
vect_list
```

```
len(vect_list) # 벡터로 더해서 나눠서 평균낸 것
```

```
sim_mat = np.zeros([len(vect_list), len(vect_list)])
for i in range(len(vect_list)):
    for j in range(len(vect_list)):
        sim_mat[i][j] = cosine_similarity(vect_list[i].reshape(1, embedding_dim), \
                                            vect_list[j].reshape(1, embedding_dim))[0,0]
        # 12*12의 유사도 행렬이 cosine_similarity(코사인 유사도)로 계산되어 출력됨
sim_mat
                                
```

```
import networkx as nx
import matplotlib.pyplot as plt

def draw_graphs(sim_matrix):
  # 유사도 행렬을 네트워크 그래프로 변환합니다.
  nx_graph = nx.from_numpy_array(sim_matrix)
  
  # 그래프 시각화를 위한 새로운 창을 생성합니다.
  plt.figure(figsize=(10, 10))
  
  # 노드의 위치를 스프링 레이아웃 알고리즘을 사용하여 정의합니다.
  pos = nx.spring_layout(nx_graph)
  
  # 노드와 라벨을 함께 그립니다.
  nx.draw(nx_graph, with_labels=True, font_weight='bold')
  
  # 네트워크 그래프의 엣지 라벨을 그립니다.
  edge_labels = {(u, v): round(sim_matrix[u][v], 2) for u, v in nx_graph.edges}
  nx.draw_networkx_edge_labels(nx_graph, pos, edge_labels=edge_labels, font_color='red')
  
  # 그래프를 출력합니다.
  plt.show()
```

```
def calculate_score(sim_matrix):
    nx_graph = nx.from_numpy_array(sim_matrix)
    scores = nx.pagerank(nx_graph)
    return scores
```
