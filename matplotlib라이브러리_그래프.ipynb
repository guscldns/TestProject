```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips") # tips 데이터(음식점데이터) 가져오기
tips
```

```python
fig = plt.figure() # 그림을 그리기 위한 도화지를 한장 준비해서 fig 변수에 할당

ax1 = fig.add_subplot(1,1,1) # 1행 1열 1번째 자리
ax1.hist(tips['total_bill'], bins=10)
# tips의 total bill이라는 값으로 막대그래프(hist) 그리기
# bins : 그래프에서 보여주는 막대기 개수
```

```python
ax1.hist(tips['total_bill'], bins=20)

fig # 그래프 보여줘

# 앞에 그린 그래프와 겹쳐서 보여줌
# 도화지 설정하고 그림 그리면 지워지는 게 아니라 그 위에 덧칠
```

```python
ax1.hist(tips['total_bill'], bins=30)
fig
```

```python
fig = plt.figure() # 그림을 그리기 위한 도화지를 한장 준비해서 fig 변수에 할당

ax1 = fig.add_subplot(1,1,1) 

ax1.scatter(tips['total_bill'], tips['tip'], c=tips['size'], s=tips['size']*20, alpha=0.5) 
# scatter(산점도) 그림함수는 X축 Y축 필요 
# X축 : total_bill / Y축 : tip
# c : 컬러 / size : 색변경 기준 
# s : 사이즈 / size : 크기변경 기준  
# * 숫자 : 점 크기에 숫자 곱해서 더 크게 보이게 함
# alpha : 투명도 설정


# 제목 설정
ax1.set_xlabel("Total Bill") # x축 제목
ax1.set_ylabel("Tip") # y축 제목
ax1.set_title("SCATTER") # 그래프 제목
```
