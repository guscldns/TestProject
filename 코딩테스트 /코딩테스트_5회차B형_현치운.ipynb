## 1. 부족한 금액 계산하기


image.png
image.png

```python
import time

start = time.time()


def solution(price, money, count):
# 답안 작성 부분 ===============

    for i in range(1, count+1):
        money -= price*i

    if (price-money) < 0:
        return 0
    else:
        return abs(money)

# ==============================

    # return answer

print(solution(3, 20, 4))


end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 2. 시저 암호

```python
import time

start = time.time()


def solution(s, n):
# 답안 작성 부분 ===============
    answer = ""
    for i in range(len(s)):
        if s[i] != ' ':
            a = ord(s[i])
            if a == 90 or a == 122:
                a -= 26
            answer += chr(a+n)
        else:
            answer += s[i]

# ==============================

    return answer

print(solution("AB",	1))
print(solution("z",	1))
print(solution("a B z",	4))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 3. 숫자 문자열과 영단어 - 2021 카카오 채용연계형 인턴십

```python
import time

start = time.time()


def solution(s):
# 답안 작성 부분 ===============
    num = {"zero":'0', "one":'1', "two":'2', "three":'3', "four":'4', "five":'5', "six":'6', "seven":'7', "eight":'8', "nine":'9' }

    for i in num:
        if i in s:
            s = s.replace(i, num[i])

    return int(s)
# ==============================

    # return answer


end = time.time()

print(solution("one4seveneight"))
print(solution("23four5six7"))
print(solution("2three45sixseven"))
print(solution("123"))

print("작동시간:", f"{end - start:.5f} sec")
```

## 4. 기사단원의 무기

```python
import time

start = time.time()


def solution(number, limit, power):
# 답안 작성 부분 ===============

    a = [1]
    for i in range(2, number+1):
        b = 0
        # for j in range(1, i+1):
        for j in range(1, int(i**0.5)+1):
            if i % j == 0:
                b += 1
                if (j**2) != i :
                    b += 1
        if b > limit:
            a.append(power)
        else:
            a.append(b)

# ==============================

    return sum(a)

print(solution(5, 3, 2))
print(solution(10, 3, 2))
end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
