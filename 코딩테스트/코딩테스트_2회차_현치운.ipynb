## 1. 피자 나눠 먹기 (1)

```python
import time

start = time.time()


def solution(n):
# 답안 작성 부분 ===============
    answer = n//7 if n%7 == 0 else (n//7)+1
# ==============================

    return answer

print(solution(7))
print(solution(1))
print(solution(15))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time

start = time.time()


def solution(dot):
# 답안 작성 부분 ===============
    if dot[0] > 0 and dot[1]> 0 :
        answer = 1
    elif dot[0] < 0 and dot[1]> 0 :
        answer = 2
    elif dot[0] > 0 and dot[1]< 0 :
        answer = 3
    elif dot[0] < 0 and dot[1]< 0 :
        answer = 4
# ==============================

    return answer

print(solution([2,4]))
print(solution([-7,9]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 3. 중복된 문자 제거


image.png

```python
import time

start = time.time()


def solution(my_string):
# 답안 작성 부분 ===============
    answer = []
    for i in list(my_string):
        if i not in answer :
            answer.append(i)
# ==============================

    return ''.join(answer)

print(solution("people"))
print(solution("We are the world"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time

start = time.time()


def solution(id_pw, db):
# 답안 작성 부분 ===============
    for i in range(len(db)):

        if id_pw[0] == db[i][0] and id_pw[1] == db[i][1]:
            answer =  "login"

        elif id_pw[0] == db[i][0] and id_pw[1] != db[i][1]:
            answer =  "wrong pw"

        else:
            answer =  "fail"
# ==============================

    return answer

print(solution(["meosseugi", "1234"],  [["rardss", "123"], ["yyoom", "1234"], ["meosseugi", "1234"]]))
print(solution(["programmer01", "15789"], [["programmer02", "111111"], ["programmer00", "134"], ["programmer01", "1145"]] ))
print(solution(["rabbit04", "98761"], 	[["jaja11", "98761"], ["krong0313", "29440"], ["rabbit00", "111333"]]  ))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 5. 멘토 제안 문제: 아래 2개 중 1개 선택 (두개다 풀어도 1개의 점수)

### 1. 기초 DNN 구현 (인공지능 코딩테스트 기초 대비, 넘파이 외 라이브러리 사용 불가능)
activation function은 시그모이드이며 최종 아웃풋의 출력값을 구하시오

image.png

```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))


def identity_function(x):
    return x

X = np.array([1.0, 0.5])
W1 = np.array([[0.1, 0.3, 0.5],[0.2, 0.4, 0.6]])
B1 = np.array([0.1, 0.2, 0.3])

A1 = np.dot(X, W1) + B1

Z1 = sigmoid(A1)  # 함수와 괄호안에 답 넣기

W2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
B2 = np.array([0.1, 0.2])

A2 =  np.dot(Z1, W2) + B2  # 수식
Z2 = sigmoid(A2)  # 함수와 괄호안에 답 넣기



W3 = np.array([[0.1, 0.3], [0.2, 0.4]])
B3 = np.array([0.1, 0.2])

A3 = np.dot(Z2, W3) + B3  # 수식
Y = identity_function(A3)
print(Y)     # array([0.31682708, 0.69627909]) 나와야함.
```

## 2. 영어가 싫어요

```python
import time

start = time.time()


def solution(numbers):
# 답안 작성 부분 ===============
    num = {"zero": '0', "one":'1', "two":'2', 'three':'3', 'four':'4', 'five':'5', 'six':'6', 'seven':'7', 'eight':'8', 'nine':'9', 'ten':'10'}

    for k,v in num.items():
        numbers = numbers.replace(k,v)
    numbers = int(numbers)
# ==============================

    return numbers

print(solution("onetwothreefourfivesixseveneightnine"))
print(solution("onefourzerosixseven"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
