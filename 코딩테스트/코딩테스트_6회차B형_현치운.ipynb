```python
import time

start = time.time()


def solution(brown, yellow):
# 답안 작성 부분 ===============
    a = int((brown+yellow)**0.5)
    answer = [int((brown+yellow)/a), a]
# ==============================

    return answer

end = time.time()

print(solution(10, 2))
print(solution(8, 1))
print(solution(24, 24))

print("작동시간:", f"{end - start:.5f} sec")
```

## 2. 구명보트


image.png

```python
import time

start = time.time()

def solution(people, limit):
# 답안 작성 부분 ===============
    p = len(people)
    people = sorted(people)
    answer = 0

    while p > 0 :
        if people[0] + people[-1] == limit:
            people = people[:-1]
            if len(people) == 0:
                answer += 1
                break
            people = people[1:]
            answer += 1
            p -= 2
        else:
            people = people[:-1]
            answer += 1
            p-=1

# ==============================

    return answer

print(solution([70, 50, 80, 50], 100))
print(solution([70, 80, 50],	100))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 3. N개의 최소공배수

image.png

```python
import time

start = time.time()


def solution(arr):
# 답안 작성 부분 ===============
    answer = 1
    for i in range(len(arr)):
        m = min(arr)
        if arr[i] != m and arr[i]%m == 0:
            arr[i] = arr[i]//m
        answer = answer * arr[i]

# ==============================

    return answer

print(solution([2,6,8,14]))
print(solution([1,2,3]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 4. 귤 고르기

```python
import time

start = time.time()


def solution(k, tangerine):
# 답안 작성 부분 ===============
    answer = 0
    a = sorted(list(set(tangerine)))
    b = {o:tangerine.count(o) for o in a}
    c = sorted(b.items(), key=lambda x:x[1], reverse=True)

    for i in range(len(c)):
        if k - c[i][1] >= 0:
            k = k-c[i][1]
            answer += 1
# ==============================

    return answer

print(solution(6,	[1, 3, 2, 5, 4, 5, 2, 3]))
print(solution(4,	[1, 3, 2, 5, 4, 5, 2, 3]))
print(solution(2, [1, 1, 1, 1, 2, 2, 2, 3]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time

start = time.time()


def solution(s):
# 답안 작성 부분 ===============
    answer = 0
    for i in range(len(s)):
        if i == 0:
            a = s
        else:
            a = s[-i:] + s[:-i]
        for i in a:
            a= a.replace("[]", '')
            a= a.replace("()", '')
            a= a.replace("{}", '')

        if len(a)== 0:
            answer +=1
# ==============================

    return answer

print(solution("[](){}"))
print(solution("}]()[{"))
print(solution("[)(]"))
print(solution("}}}"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
