# 0. 푸는 방법 설명

문제 설명

0에서 n까지 합을 return 하도록 함수를 완성해주세요

```python
import time

start = time.time()

# 답안 작성 부분 ===============
def solution(n):
    answer = 0
    for i in range(0,n+1):
        answer += i
    return answer

print(solution(10))
print(solution(5))

# ==============================

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time

start = time.time()

# 답안 작성 부분 ===============
def solution(num1, num2):
    answer = num1//num2

    return answer

print(solution(10, 5))
print(solution(7, 2))

# ==============================

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

# 2. 배열 자르기

image.png

```python
import time

start = time.time()

# 답안 작성 부분 ===============
def solution(numbers, num1, num2):
    answer = numbers[num1:num2+1]

    return answer

print(solution([1,2,3,4,5], 1, 3))
print(solution([1,3,5], 1, 2))

# ==============================

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

# 4. 콜라츠 추측

```python
import time

start = time.time()

# 답안 작성 부분 ===============
def solution(n):
    answer = 0
    while n > 1:
        if n %2 == 0:
            n = n /2

        else:
           n = (n * 3)+1
        answer +=1

    if answer >= 500:
        answer = -1

    return answer

print(solution(6))
print(solution(16))
print(solution(626331))

# ==============================

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

# 5.  휴대폰 번호 가리기

image.png

```python
import time

start = time.time()

# 답안 작성 부분 ===============
def solution(phone_number):
    answer = len(phone_number[:-4])* '*' + phone_number[-4:]

    return answer

print(solution("01033334444"))
print(solution("027778888"))

# ==============================

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
