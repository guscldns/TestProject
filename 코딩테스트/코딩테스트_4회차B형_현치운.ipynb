## 1. 기사단원의 무기


image.png
image.png

```python
import time

start = time.time()


def solution(number, limit, power):
# 답안 작성 부분 ===============
    num = list(range(1, number+1))
    pow = []
    result = 0

    for i in num:
        for n in range(1,i+1):
            if i%n == 0:
                result += 1
        if result > limit:
            pow.append(power)
            result = 0
        else:
            pow.append(result)
            result = 0
# ==============================

    return sum(pow)

print(solution(5, 3, 2))
print(solution(10, 3, 2))
end = time.time()

# print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time

start = time.time()


def solution(babbling):
# 답안 작성 부분 ===============
    bb = {"aya":"", "ye":"", "woo":"", "ma": ""}
    for i in range(len(babbling)):
        for j in bb:
            if j in babbling[i]:
                babbling[i] = babbling[i].replace(j, bb[j],1)

# ==============================

    return babbling.count("")

print(solution(["aya", "yee", "u", "maa"]))
print(solution(["ayaye", "uuu", "yeye", "yemawoo", "ayaayaa"]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 3. 최댓값과 최솟값


image.png

```python
import time

start = time.time()

def solution(s):
# 답안 작성 부분 ===============
    s = [int(i) for i in s.split(" ")]
    answer = f"{min(s)} {max(s)}"

# ==============================

    return answer

print(solution("1 2 3 4"))
print(solution("-1 -2 -3 -4"))
print(solution("-1 -1"))


end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 4. JadenCase 문자열 만들기

image.png

```python
import time

start = time.time()


def solution(s):
# 답안 작성 부분 ===============
    s = s.split(" ")
    answer =[]
    for i in s:
        if i:
            answer.append(i[0].upper()+i[1:].lower())
        else:
            answer.append(i)

# ==============================

    return ' '.join(answer)

print(solution("3people unFollowed me"))
print(solution("for the last week"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
