## 1. 부족한 금액 계산하기


image.png
image.png

```python
import time

start = time.time()


def solution(price, money, count):
# 답안 작성 부분 ===============
    n_price = (price *(1+count)) * count//2

    if n_price-money > 0:
        answer = n_price-money
    else:
        answer = 0
# ==============================

    return answer

print(solution(3, 20, 4))


end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 2. 행렬의 덧셈

![image.png]()image.png

```python
import time

start = time.time()


def solution(arr1, arr2):
# 답안 작성 부분 ===============
    for i in range(len(arr1)):
        for j in range(len(arr1[i])):

           arr1[i][j] = arr1[i][j] + arr2[i][j]

# ==============================

    return arr1

print(solution([[1,2],[2,3]],	[[3,4],[5,6]]))
print(solution([[1],[2]],	[[3],[4]]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 3. 최대공약수와 최소공배수


```python
import time

start = time.time()

def solution(n, m):
# 답안 작성 부분 ===============
    a = min(n,m)
    b = max(n,m)
    result = []
    for i in range(2,a+1):
        if a%i == 0 and b%i == 0:
            result.append(i)
            break
    if len(result) == 0:
        result.append(1)

    if b%a == 0:
        result.append(b)
    else:
        result.append(a*b)

    answer = [min(result), max(result)]

# ==============================

    return answer

print(solution(3, 12))
print(solution(2, 5))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time

start = time.time()


def solution(s, n):
# 답안 작성 부분 ===============
    s = list(s)
    for i in range(len(s)):
        if s[i].isalpha():
            a = ord(s[i])
            if a == 122 or a == 90:
                a = a-26
            s[i] = chr(a+n)
    answer = ''.join(s)


# ==============================

    return answer

print(solution("AB",	1))
print(solution("z",	1))
print(solution("a B z",	4))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 5. 숫자 문자열과 영단어 - 2021 카카오 채용연계형 인턴십

```python
import time

start = time.time()


def solution(s):
# 답안 작성 부분 ===============
    number = {"one":"1", "two":"2", "three": "3", "four":"4", "five": "5", "six":"6", "seven": "7", "eight":"8", "nine":"9", "zero":"0"}
    for i in number:
        if i in s :
            s = s.replace(i,number[i])
# ==============================

    return int(s)


end = time.time()

print(solution("one4seveneight"))
print(solution("23four5six7"))
print(solution("2three45sixseven"))
print(solution("123"))

print("작동시간:", f"{end - start:.5f} sec")
```
