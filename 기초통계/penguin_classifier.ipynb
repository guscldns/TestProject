## 2. 분류기 만들기

- 아래 그림을 참조하여 코드의 빈칸을 채우세요.



image.png

````python
def clf(X):
    weight = [
                [-9.04912495e-01,  2.14668886e+00,  6.95066507e-04],
                [ 6.79854815e-01, -5.26267979e-01, -4.93941073e-03],
                [ 2.25057680e-01, -1.62042089e+00,  4.24434423e-03]
            ]
    intercept = [ 0.1074035 , -0.07936444, -0.02803905]

    # 모든 샘플의 스코어를 저장할 리스트를 초기화 합니다.
    scores = []

    # X에 저장된 펭귄 샘플을 하나씩 x로 가져오기 위해 X에 대해서 for문을 돌립니다.
    for x in X:
        # 첫반복에서 x: [43.3, 14.0, 4575.0]
        # 샘플 하나에 대한 스코어 값을 저장할 리스트를 초기화 합니다.
        # 이 리스트에는 현재 샘플 x가 Adelie, Chinstrap, Gentoo일 것 같은 점수가
        # 저장됩니다. 즉 이 리스트에는 숫자 3개가 저장되어야 합니다.
        model_score = []

        # model_score에 저장될 숫자 3개를 계산하기 위해 바깥 for문을 3번 돌립니다.
        for i in range(3):
            # Adelie, Chinstrap, Gentoo에 대한 가중치값 3개를 가져옵니다.
            w = weight[i] # 첫반복에서 [-9.04912495e-01,  2.14668886e+00,  6.95066507e-04]
            # z = w[0]*x[0] + w[1]*x[1] + w[2]*x[2] 을 계산하기 위해
            # z를 초기화 하고
            z = 0.0
            # x와 w에 들어있는 숫자 3개를 각각 곱하고 모두 더하기 위해 안쪽 for문을 3번 돌립니다.
            for j in range(3):
                # z = w[0]*x[0] + w[1]*x[1] + w[2]*x[2]에 각 항을 순서대로 계산하여
                # z에 누적합니다.
                z += w[j]*x[j] ## None를 적절한 코드로 바꾸세요. ##

            # 마지막으로 intercept를 z에 더합니다. intercept에는 숫자 3개가 들어있는데
            # 어떤 숫자를 더해야하는지 인덱스를 결정하세요.
            z += intercept[i] ## None를 적절한 코드로 바꾸세요. ##
            # 계산을 마친 z를 model_score에 저장합니다.
            model_score.append(z) ## None를 적절한 코드로 바꾸세요. ##

        # 이전 코드가 제대로 실행이 되었다면 이 시점에서 model_score에는 숫자 3개가 들었습니다.
        # 모든 스코어를 저장할 scores 리스트에 model_score를 저장합니다.
        scores.append(model_score) ## None를 적절한 코드로 바꾸세요. ##

    return scores

````
