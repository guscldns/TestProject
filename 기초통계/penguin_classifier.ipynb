````python
def clf(X):
    weight = [
                [-9.04912495e-01,  2.14668886e+00,  6.95066507e-04],
                [ 6.79854815e-01, -5.26267979e-01, -4.93941073e-03],
                [ 2.25057680e-01, -1.62042089e+00,  4.24434423e-03]
            ]
    intercept = [ 0.1074035 , -0.07936444, -0.02803905]

    #       .
    scores = []

    # X     x   X  for .
    for x in X:
        #  x: [43.3, 14.0, 4575.0]
        #         .
        #     x Adelie, Chinstrap, Gentoo   
        # .     3  .
        model_score = []

        # model_score   3    for 3 .
        for i in range(3):
            # Adelie, Chinstrap, Gentoo   3 .
            w = weight[i] #  [-9.04912495e-01,  2.14668886e+00,  6.95066507e-04]
            # z = w[0]*x[0] + w[1]*x[1] + w[2]*x[2]   
            # z  
            z = 0.0
            # x w   3       for 3 .
            for j in range(3):
                # z = w[0]*x[0] + w[1]*x[1] + w[2]*x[2]    
                # z .
                z += w[j]*x[j] ## None   . ##

            #  intercept z . intercept  3 
            #     .
            z += intercept[i] ## None   . ##
            #   z model_score .
            model_score.append(z) ## None   . ##

        #        model_score  3 .
        #    scores  model_score .
        scores.append(model_score) ## None   . ##

    return scores

````
