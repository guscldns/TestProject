##### 폴리곤 y 데이터

```python
# 폴리곤 좌표
def convert_to_segmentation(poly, image_width, image_height):
    segmentation = []
    for key, value in poly.items():
        # if isinstance(value, list) and value: # isinstance: 형태가 일치하는지 여부(T/F)
        if len(value[0]) > 1:
            for sub_dict in value:
                test = []
                for i in range(1,int(len(sub_dict)//2) +1):
                    x_key = sub_dict[f'X좌표{i}'] / image_width
                    y_key = sub_dict[f'Y좌표{i}'] / image_height
                    x_value = x_key
                    y_value = y_key
                    test.extend([x_value, y_value]) # extend: 가장 바깥쪽 iterable의 모든 항목 추가
                segmentation.append(test)
    return segmentation
```

```python
def read_txt_file(file_path):
    """
    텍스트 파일에서 특정 레이블(target_label)의 데이터만 추출하는 함수
    :param file_path: 텍스트 파일 경로
    :param target_label: 추출할 레이블
    :return: target_label에 해당하는 첫 번째 x_center, y_center, 너비(w), 높이(h) 값
    """
    with open(txt_path, 'r') as f:
        lines = f.readlines()

    x_center = None
    y_center = None
    width = None
    height = None

    for line in lines:
        label, x_center, y_center, w, h = line.strip().split(' ')
        label = int(label)
        if label == 6:
            x_center = float(x_center)
            y_center = float(y_center)
            width = float(w)
            height = float(h)
            break

    return x_center, y_center, width, height

def ratio_to_coordinates(x, y, w, h, image_width, image_height):
    x_coord = x * image_width
    y_coord = y * image_height
    width_coord = w * image_width
    height_coord = h * image_height

    x_coord = max(0, min(x_coord, image_width))
    y_coord = max(0, min(y_coord, image_height))
    width_coord = max(0, min(width_coord, image_width - x_coord))
    height_coord = max(0, min(height_coord, image_height - y_coord))

    return int(x_coord), int(y_coord), int(width_coord), int(height_coord)
```

```python
import os
import cv2

def ratio_to_coordinates(x, y, w, h, image_width, image_height):
    x_coord = x * image_width
    y_coord = y * image_height
    width_coord = w * image_width
    height_coord = h * image_height

    x_coord = max(0, min(x_coord, image_width))
    y_coord = max(0, min(y_coord, image_height))
    width_coord = max(0, min(width_coord, image_width - x_coord))
    height_coord = max(0, min(height_coord, image_height - y_coord))

    return int(x_coord), int(y_coord), int(width_coord), int(height_coord)

def cutmix(image_folder, txt_folder, output_folder, imsize):
    image_files = os.listdir(image_folder)

    for image_file in image_files:
        # Check if the file is a jpg image
        if not image_file.lower().endswith(".jpg"):
            continue

        # 이미지 파일의 이름과 확장자를 분리하여 이미지 파일의 경로를 지정
        image_path = os.path.join(image_folder, image_file)
        image = cv2.imread(image_path, cv2.IMREAD_COLOR)

        if image is None:
            print(f"Error loading image: {image_path}")
            continue

        # 이미지 파일의 이름과 같은 이름의 txt 파일을 찾아서 txt 파일의 경로를 지정
        txt_file = os.path.splitext(image_file)[0] + ".txt"
        txt_path = os.path.join(txt_folder, txt_file)

        with open(txt_path, 'r') as f:
            for line in f:
                label, x, y, w, h = map(float, line.strip().split())
                label = int(label)

                if label == 6:
                    image_height, image_width = image.shape[:2]
                    x_coord, y_coord, width_coord, height_coord = ratio_to_coordinates(x, y, w, h, image_width, image_height)

                    if width_coord > 0 and height_coord > 0:
                        cut_image = image[y_coord:y_coord+height_coord, x_coord:x_coord+width_coord]

                        another_image_path = os.path.join(image_folder, "another_image.jpg")
                        another_image = cv2.imread(another_image_path, cv2.IMREAD_COLOR)

                        if another_image is None:
                            print(f"Error loading another image: {another_image_path}")
                            continue

                        another_image[y_coord:y_coord+height_coord, x_coord:x_coord+width_coord] = cut_image

                        result_image_path = os.path.join(output_folder, f"{image_file}_cutmix.jpg")
                        cv2.imwrite(result_image_path, another_image)

image_folder = "/content/cutmix" # 이미지가 저장된 폴더 경로
txt_folder = "/content/cutmix" # txt 파일이 저장된 폴더 경로
output_folder = "/content/cutmix_results1" # cutmix 결과 이미지를 저장할 폴더 경로
imsize = 512 # cutmix 결과 이미지의 크기

cutmix(image_folder, txt_folder, output_folder, imsize)
```
